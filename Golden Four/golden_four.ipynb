```python
import pandas as pd
import folium
geo='https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json'
data=pd.read_excel(r'C:\Users\arek\Desktop\Golden Four\data.XLSX')
data[2012]=data[2012].astype(float)
m = folium.Map(location=[0,0], zoom_start=2)

m.choropleth(
    geo_data=geo,
    name='choropleth',
    data=data,
    columns=['Country Name', 2012],
    key_on='properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Zalesienie {?$@!}'
)

folium.LayerControl().add_to(m)

m
```
